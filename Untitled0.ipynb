# Ingenieria de datos

¿que es ingenieria de datos?

Es un campo del area de sistemas que se enfoca en el diseño, desarrollo y gestion de sistemas, que recopilan y almacenan datos.

El objetivo final de un data enginner es hacer que los datos sean accesibles para que las organizaciones puedan utulizarlos para evaluar y optimizar sus rendimientos.

**Ademas es importante porque permite a las otras areas de la empresa a tomar desiciones**
image.png

# ¿Que hace un ingeniero de datos?



1.   Recopilacion de datos
##### Obtencion de datos de diversas fuentes, pueden ser bases de datos, APIS, archivos de texto, imagenes
2.   Tranformacion de datos
##### Limpieza y procesamiento de datos para asegurar su calidad y formato adecuado para el analisís
3.   Almacenamiento de datos
##### Almacenar los datos tranformados y de facil acceso
4.   Pipeline de datos
##### Creacion y mantenimiento de un flujo de trabajo que funcione de manera automatica y que llegue todo a su destino
5.   Seguridad de datos
##### Asegurar que los datos esten protegidos y que la informacion este tokenizada.

````python
import pandas as pd
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
pd.set_option('display.width', 1000)  # Establecer un ancho grande para evitar saltos de línea

# Datos ficticios de clientes
data_clientes = {
    'ID_Cliente': [1, 2, 3, 4, 5],
    'Nombre': ['Juan Pérez', 'María García', 'Carlos López', 'Ana Martínez', 'Luis Rodríguez'],
    'Email': ['juan.perez@example.com', 'maria.garcia@example.com', 'carlos.lopez@example.com', 'ana.martinez@example.com', 'luis.rodriguez@example.com']
}

# Crear el DataFrame de clientes
df_clientes = pd.DataFrame(data_clientes)

# Datos ficticios de ventas
data_ventas = {
    'ID_Cliente': [1, 2, 1, 3, 4],  # Relaciona las ventas con los clientes
    'Fecha': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Producto': ['Producto A', 'Producto B', 'Producto A', 'Producto C', 'Producto B'],
    'Cantidad': [10, 5, 8, 12, 7],
    'Precio_Unitario': [15.0, 20.0, 15.0, 30.0, 20.0],
    'Total_Venta': [150.0, 100.0, 120.0, 360.0, 140.0]
}

# Crear el DataFrame de ventas
df_ventas = pd.DataFrame(data_ventas)

# Convertir la columna 'Fecha' a tipo datetime
df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'])

# Mostrar los DataFrames
print("Tabla de Clientes")
print(df_clientes)
print("\nTabla de Ventas")
print(df_ventas)
````

````python
df_joined = pd.merge(df_ventas, df_clientes, on='ID_Cliente', how='inner')
````
