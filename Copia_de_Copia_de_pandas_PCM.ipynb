#Libreria Pandas
Pandas es una librería que permite cargar un conjunto de datos (csv, xlsx, etc.), explorar el conjunto de datos, realizar operaciones y analizar los datos.

Los elementos básicos en Pandas son las **series** y los **dataframes**

```python
import pandas as pd
import numpy as np

#Crear una serie
array = np.random.randint(1,50,(50,))
print(array)
serie = pd.Series(array)
print(serie)
```

```python
#otro ejemplo (ahora podremos indexar cada elemento)
ventas = pd.Series(
    [10,30,30,49,20,20,10],
    index=['Tienda Pepe','Market Lola','Ciencia Stores','Python bodeguita','La esquina Market','Tiendita','Supermarket R']
)
print(ventas)
```

```python
#Podemos colocar un nombre a nuestro serie
ventas.name = "Ventas de junio 2023 (en soles)"
print(ventas)
```

```python
#Colocando un indice a los indices
ventas.index.name = "Tiendas"
print(ventas)
```

```python
print(ventas.shape)
```

#Dataframe
Es una estructura de datos bidimensional heterogénea(se admite amplia variedad de tipos de datos). Las columnas y filas están indizadas de modo que el acceso a la información sea efectivo. Es la estructura principal de Pandas.


```python
#Creando un dataframe a partir de un diccionario
trabajadores = { "Nombres":['Juan','Carlos','Maria','Carla','Mike'],
                "Apellidos":['Perez','Quispe','Torres','Alvarez','Rodriguez'],
                 "Cargo":['Asistente','Tecnico','Gerente','Asistente','Asesor'],
                 "Sueldo":[2000,1500,5000,2500,3000],
                 "Años de experiencia":[3,3,10,5,6]
}
#print(trabajadores)

trabajadoresdf = pd.DataFrame(trabajadores)
trabajadoresdf
```

```python
#Esto permite acceder a cada columna
trabajadoresdf['Cargo']
```

```python
#Una columna puede convertirse en los índices del dataframe
trabajadoresdf.index = trabajadoresdf['Nombres']
trabajadoresdf
```

```python
trabajadoresdf.info()
```

```python
trabajadoresdf.describe()
```

#Acceder a elementos
Se puede acceder a los elementos de un dataframe de modo similar al de una lista. Sin embargo, existen otros comandos y estos son **loc()** y **iloc()**.

```python
#Accediendo de modo similar a una lista
trabajadoresdf['Apellidos'][2]
```

```python
#Podemos colocar el nombre de la columna y de fila
trabajadoresdf['Cargo']['Juan']
```

**loc()** permite acceder a los elementos por nombre de columna y fila, y también permite hacer consultas lógicas para filtrar datos.

```python
trabajadoresdf.loc['Maria']
trabajadoresdf.loc['Maria']['Años de experiencia']
```

```python
#Los dos puntos sirven para abarcar un rango
trabajadoresdf.loc['Carlos':'Carla']
trabajadoresdf.loc['Carlos':'Carla','Cargo']
```

```python
#filtrar data
trabajadoresdf.loc[trabajadoresdf['Sueldo']<2200]
trabajadoresdf.loc[trabajadoresdf['Años de experiencia']>5]
```

**iloc()** admite los índices numéricos para acceder a los datos, o fracciones de este.

```python
trabajadoresdf.iloc[0]
trabajadoresdf.iloc[0:3]
trabajadoresdf.iloc[0:3,2]
```


#Operaciones entre columnas

Podemos realizar operaciones entre columnas, de modo que el resultado se agrege en una nueva columna.

```python
#Creamos la columna Aumento que es el 10% de Sueldo
trabajadoresdf['Aumento'] = 0.1*trabajadoresdf['Sueldo']
trabajadoresdf
```

Explores un ejemplo diferente. Se creará un conjunto de con datos aleatorios sobre masa (en kg), talla(en metros m), edad, ingresos(en soles) de un conjunto de personas. Se realizará operaciones a las columnas, y entre columnas.

```python
masa = np.random.randint(45,120,(100,))
talla = np.random.randint(150,210,(100,))
edad = np.random.randint(18,80,(100,))
ingresos = np.random.randint(1000,3000,(100,))

df = pd.DataFrame(
    {'Masa':masa,'Talla':talla,'Edad':edad,'Ingresos':ingresos}
)
df
```

apply() nos permite definir una función y aplicarla a toda una columna. Aquí se define la funcion *adulto mayor* que retorna True si la edad es mayor de 60, y False en caso contrario.

```python
def adultomayor(edad):
  if edad >=60:
    return True
  else:
    return False
```

```python
df['EsAdultoMayor']=df['Edad'].apply(adultomayor)
```

```python
df
```

El Indice de Masa Corporal (IMC) es la relación entre la masa y talla. Su valor determina si una persona tiene o no sobrepeso

```python
df['IMC'] = df['Masa']/(df['Talla']/100)**2
df
```

Si el IMC es mayor de 25, se considera que la persona tiene sobrepeso. Definamos una función para esto:

```python
def sobrepeso(imc):
  if imc >=25:
    return True
  else:
    return False
```

```python
df['Sobrepeso']= df['IMC'].apply(sobrepeso)
df
```

groupby() nos permite agrupar datos por columnas, ejemplo:

```python
df.groupby(['Sobrepeso']).size()
```

```python
df.groupby(['EsAdultoMayor','Sobrepeso']).size()
```

#Gráficos con Pandas
Pandas permite realizar algunos gráficos para inspeccionar los datos a analizar. Veamos algunos ejemplos:

```python
df.groupby(['Sobrepeso']).size().plot.bar()
```

```python
df.groupby(['Sobrepeso','EsAdultoMayor']).size().plot.bar()
```

```python
df.groupby(['Sobrepeso']).size().plot.pie()
```
