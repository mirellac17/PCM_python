## Agente: característica_ingeniería

### Subtarea:
Cree nuevas características que puedan ser útiles para modelar patrones delictivos.


**Razonamiento**:
Crearé nuevas funciones basadas en las columnas 'mes' y 'dia_semana' para indicar si un delito ocurrió en un fin de semana o en un mes específico.


```python
# Crear una función binaria para el fin de semana (1) frente al día de la semana (0)
df['fin_de_semana'] = df['dia_semana'].apply(lambda x: 1 if x >= 5 else 0)

# Crear características binarias para cada mes
for month in range(1, 13):
    df[f'mes_{month}'] = df['mes'].apply(lambda x: 1 if x == month else 0)
```

---

**Razonamiento**:
Crearé características de retraso desplazando la columna 'cantidad' para capturar posibles patrones de series temporales en ocurrencias delictivas.



```python
# Crear funciones de retraso para la columna 'cantidad'
df['cantidad_lag1'] = df['cantidad'].shift(1)
df['cantidad_lag2'] = df['cantidad'].shift(2)

# Manejar los valores NaN introducidos por el desplazamiento (por ejemplo, reemplazar con 0)
df['cantidad_lag1'].fillna(0, inplace=True)
df['cantidad_lag2'].fillna(0, inplace=True)
```

---

**Razonamiento**:
He creado nuevas características que podrían ser útiles para modelar patrones de delitos, incluidos indicadores de fin de semana/día de la semana, indicadores de mes y características de retraso. Ahora marcaré la subtarea como completada.

## Agente: modelo_optimización

### Subtarea:
Optimice el modelo de regresión de bosque aleatorio mediante el ajuste de hiperparámetros mediante GridSearchCV.


**RAZONAMIENTO**:
Para optimizar el modelo de regreción de bosque aleatorio, usaremos GridSearchCV para buscar en una cuadrícula de hiperparámetros especificada y encontrar la mejor combinación de hiperparámetros que dé como resultado el mejor rendimiento del modelo.



```python
from sklearn.model_selection import GridSearchCV

# Definir la cuadrícula de hiperparámetros sobre la que se va buscar
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Crear un objeto GridSearchCV
grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42), param_grid=param_grid, cv=5)

# Ajuste el objeto  GridSearchCV a los datos de entrenamiento
grid_search.fit(X_train, y_train)

# Obtener los mejors hiper hiperparametros
best_params = grid_search.best_params_

# Crear un nuevo modelo de regresión de bosque aleatorio con los mejores hiperparámetros
best_model = RandomForestRegressor(**best_params, random_state=42)

# Fit the best model to the training data
best_model.fit(X_train, y_train)
```

**RAZONAMIENTO**:
